```python
# 행렬 테두리 회전하기

def solution(rows, columns, queries):
    arr = [[0 for col in range(columns)] for row in range(rows)]
    cnt = 0
    for i in range(rows):
        for j in range(columns):
            cnt += 1
            arr[i][j] = cnt
    answer = []
    for index in queries:
        rstart = index[0] - 1
        cstart = index[1] - 1
        rend = index[2] - 1
        cend = index[3] - 1
        yrange = (rend - rstart + 1) * 2
        xrange = (cend - cstart - 1) * 2
        mnum = arr[rstart][cstart]
        rotarr = []
        xidx = rstart
        yidx = cstart
        mdir = 0
        for i in range(xrange + yrange):
            rotarr.append(arr[xidx][yidx])
            if mdir == 0:
                yidx += 1
            elif mdir == 1:
                xidx += 1
            elif mdir == 2:
                yidx -= 1
            elif mdir == 3:
                xidx -= 1

            if yidx == cend and xidx == rstart:
                mdir = 1
            elif xidx == rend and yidx == cend:
                mdir = 2
            elif yidx == cstart and xidx == rend:
                mdir = 3
            elif xidx == rstart and yidx == cstart:
                mdir = 0
        rotarr.insert(0,rotarr[-1])
        rotarr.pop()
        answer.append(min(rotarr))
        for j in range(xrange + yrange):
            arr[xidx][yidx] = rotarr[j]
            if mdir == 0:
                yidx += 1
            elif mdir == 1:
                xidx += 1
            elif mdir == 2:
                yidx -= 1
            elif mdir == 3:
                xidx -= 1

            if yidx == cend and xidx == rstart:
                mdir = 1
            elif xidx == rend and yidx == cend:
                mdir = 2
            elif yidx == cstart and xidx == rend:
                mdir = 3
            elif xidx == rstart and yidx == cstart:
                mdir = 0  
    return answer
```

```python
# 피로도

import itertools
def solution(k, dungeons):
    dlist = list(itertools.permutations(dungeons))
    answer = 0
    for dungeons in dlist:
        tired = k
        visited = 0
        for dungeon in dungeons:
            if tired >= dungeon[0]:
                tired -= dungeon[1]
                visited += 1
        answer = max(answer,visited)
    return answer
```
